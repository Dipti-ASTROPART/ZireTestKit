```python
import uproot
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib widget

def plot_pst_data(data, channel, title, limits=None):
    plt.figure(figsize=(10, 6))
    plt.plot(data[channel], marker='o', linestyle='-', markersize=2, label=channel)
    plt.title(title)
    plt.xlabel('Sample Number')
    plt.ylabel('ADC Counts')
    plt.xlim(limits[0], limits[1]) if limits else None
    plt.grid(True)
    plt.legend()
    plt.show()

def plot_histogram(data, channel, title):
    plt.figure(figsize=(10, 6))
    plt.hist(data[channel], bins=100, alpha=0.7, label=channel)
    plt.title(title)
    plt.xlabel('ADC Counts')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
input_file = uproot.open("/Users/dipti/Work/DataDir/SiPM_Characterization/PST_SiPM_Calibration/PST_CH04/PST_Ch04_44.4V/data.root")
input_tree = input_file["output_tree"]
branch_name = ['PST_Layer1_A_HG']
data = input_tree.arrays(branch_name, library="pd")

plot_pst_data(data, branch_name, "PST Layer 1 A HG Data", limits=(3500, 7500))
# plot_histogram(data, branch_name, "PST Layer 1 A HG Histogram")

```
